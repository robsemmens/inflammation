```python
## making a function

def fahr_to_kelvin(temp):
    return ((temp -32) * (5/9)) + 273.15
```

```python
observed_temps = [10, 87, 30, 40, 61, 82, 101, 450, 27]

for i in observed_temps:
    print(round(fahr_to_kelvin(i), 2))
```

```python
def kelvin_to_celsius(temp_k):
    return temp_k - 273.15
```

```python
def fahr_to_celsius(temp_f):
        temp_k = fahr_to_kelvin(temp_f)
        return kelvin_to_celsius(temp_k)
```

```python
observed_temps = [10, 87, 30, 40, 61, 82, 101, 450, 27]

for i in observed_temps:
    print("Observed temp in C:",  round(fahr_to_celsius(i), 2))
```

```python
import glob
import numpy
import matplotlib.pyplot as plt

filenames = sorted(glob.glob("data/inflammation-*.csv"))

def analyze(f):
    data = numpy.loadtxt(fname=f, delimiter=",")
#    image = plt.imshow(data)
#    plt.show(image)
    
    fig = plt.figure(figsize=(10.0, 3.0))
    
    matrix_mean = numpy.mean(data,axis=0)
    matrix_max = numpy.max(data,axis=0)
    matrix_min = numpy.min(data,axis=0)
    
    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)
    
    axes1.set_ylabel("average")
    axes1.plot(matrix_mean)
    
    axes2.set_ylabel("max")
    axes2.plot(matrix_max)
    
    axes3.set_ylabel("min")
    axes3.plot(matrix_min)
    
    plt.show()
    return matrix_mean, matrix_max, matrix_min
```

```python

def check_min(min_vector):
    if numpy.std(min_vector) == 0:
        print("WARNING:  Suspicious minima, motha fucka!!!")
```

```python
all_results = []
for f in filenames:
    print(f)
    matrix_mean, matrix_max, matrix_min = analyze(f)
    check_min(matrix_min)
    all_results.append([f, matrix_mean, matrix_max, matrix_min])
```
